```python
from PIL import Image
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# List Items in ./train/image/
image_dir = './train/image/'
image_list = os.listdir(image_dir)

# Segmentation DIrectory
seg_dir = './train/seg/'
seg_list = os.listdir(seg_dir)

# Item to load
item = 1234
```

```python
# Print the length of the image list
print(len(image_list))
print(len(os.listdir("./val/image/")))
```

```python
# Open image from ./train/image/ and find corresponding image in ./train/seg/
image = Image.open(image_dir + image_list[item])
seg = Image.open(seg_dir + seg_list[item])
```

```python
# Load images side by side in matplotlib, single figure
fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[1].imshow(seg)
```

```python
# Convert segmented image to numpy array
seg_array = np.array(seg)

# Create a bar graph of different values that appears in the segmented image
seg_df = pd.DataFrame(seg_array)

# Count the number of each value in the segmented image
seg_count = seg_df.stack().value_counts()

# Plot the bar graph
seg_count.plot(kind='bar')
```
